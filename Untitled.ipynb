```python
def infer_action_from_observation(previous_pose:RigidTransform, new_pose:RigidTransform, previous_gripper_is_open:bool, new_gripper_is_open:bool):
    '''
    converts from rigidtransform pose to the UR format pose (x,y,z,rx,ry,rz)
    # gripper_action = 1 if gripper_is_open changes from False to True
    # gripper_action = -1 if gripper_is_open changes from True to False
    # no change = 0 if gripper_is_open does not change
    '''
    previous_state = previous_pose.translation.tolist() + previous_pose.axis_angle.tolist()
    new_state = new_pose.translation.tolist() + new_pose.axis_angle.tolist()
    delta_state = list(np.array(new_state) - np.array(previous_state))

    gripper_action = int(new_gripper_is_open) - int(previous_gripper_is_open)
    return delta_state + [gripper_action] + [0] # 0 is the termination action
```

```python
action_traj2 = []
for i in range(len(state_traj["poses"])-1):
    action_traj2.append(infer_action_from_observation(state_traj["poses"][i], state_traj["poses"][i+1], state_traj["gripper_is_open"][i], state_traj["gripper_is_open"][i+1]))
action_traj2.append([0,0,0,0,0,0,0,1])
```
